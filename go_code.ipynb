## Auto-Commit Setup

Run the cell below to enable auto-commit after each code execution.

```go
%%
import (
    "os/exec"
    "fmt"
    "time"
)

func autoCommit() {
    time.Sleep(500 * time.Millisecond)
    
    // Git add
    exec.Command("git", "add", "go_code.ipynb").Run()
    
    // Git commit
    msg := fmt.Sprintf("Update notebook: %s", time.Now().Format("2006-01-02 15:04:05"))
    exec.Command("git", "commit", "-m", msg, "--author=Geltrax69 <10lalitsingh01@gmail.com>").Run()
    
    // Git push
    exec.Command("git", "push", "origin", "main").Run()
    
    fmt.Println("âœ“ Auto-committed and pushed to GitHub")
}

// Register to run after each cell
defer autoCommit()
```
